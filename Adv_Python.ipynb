```python
def even_odd_sum(lst):
   even_sum = 0
   odd_sum = 0
   for i in lst:
      if i%2==0:
        even_sum+= i
      else:
        odd_sum+= i
   return even_sum,odd_sum

sum1,sum2=even_odd_sum([1,2,3,4,5,3,5,6,77,5,45,32,12])
print(sum1,sum2)
```

```python
class Car:
  #constructor
  def __init__(self,window,door,enginetype):
    self.windows = window
    self.doors = door
    self.enginetype = enginetype

  def self_driving(self):
    return "this is a {} car".format(self.enginetype)

car1 = Car(4,2,"petrol")
car2 = Car(6,4,"diesel")

print(car1.windows)
print(car2.enginetype)
car1.self_driving()



```

```python
class Animal:
    def __init__(self,*args):
        if len(args)==1:
            self.name = args[0]
        elif len(args)==2:
            self.name = args[0]
            self.species = args[1]
        elif len(args)==3:
            self.name = args[0]
            self.species = args[1]
            self.age = args[2]

    def make_sound(self, sound):
        return "The animal is {} and says {}".format(self.name, sound)

dog=Animal("dog","mammals",17)
print(dog.name)
print(dog.species)
print(dog.age)
print(dog.make_sound("woof woof"))
```

```python
class Car:
  def __init__(self,window,door,enginetype):
    self.windows = window
    self.doors = door
    self.enginetype = enginetype

  def driving(self):
    print("Car is used for driving")

class audi(Car):
  def __init__(self,window,door,enginetype,horsepower):
    super().__init__(window,door,enginetype)
    self.horsepower = horsepower
audiq7 = audi(4,4,"petrol",2000)
print(audiq7.horsepower)
print(audiq7.enginetype)
audiq7.driving()
```

**Lambda**
In Python, a lambda function (also called an anonymous function) is a small, anonymous function that can be defined in a single line of code without a name. It is useful when we need a simple function that we don't want to define explicitly using the def keyword. The basic syntax for a lambda function in Python is:

lambda arguments: expression
Here, arguments refer to the input arguments for the function and expression is a single expression that gets evaluated and returned as the result of the function. The result of the expression is automatically returned by the lambda function, so there's no need to use the return statement

```python
f=lambda x,y:x+y
f
```

```python
f(5,6)
```

```python
## simnples examples
multiply=lambda x,y:x*y
multiply(3,4)
```

```python

## Return the length of a string
string_length=lambda s:len(s)
string_length("Krish Naik")
```

```python

## Convert a list of integers to their corresponding square values:
numbers=[1,2,3,4,5,6]
squares=list(map(lambda x:x**2,numbers))
print(squares)
```

```python
## Filter out even numbers from a list:

numbers=[1,2,3,4,5,6]
list(filter(lambda x:x%2==0,numbers))
```

```python
f=lambda x:x%2==0
f(3)
```

```python
## Sort a list of strings based on their alphabetical characters and  length:
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
sorted(fruits,key=lambda x:len(x))
```

```python
## complex examples
## Sorting a list of dictionaries based on a specific key
people = [
    {'name': 'Alice', 'age': 25, 'occupation': 'Engineer'},
    {'name': 'Bob', 'age': 30, 'occupation': 'Manager'},
    {'name': 'Charlie', 'age': 22, 'occupation': 'Intern'},
    {'name': 'Dave', 'age': 27, 'occupation': 'Designer'},
]
sorted(people,key=lambda x:(x['age']))
```

```python
## Finding the maximum value in a dictionary
data = {'a': 10, 'b': 20, 'c': 5, 'd': 15}
max(data,key=lambda x:data[x])
```

```python
## Grouping a list of strings based on their first letter
from itertools import groupby

words = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig']

groups = groupby(sorted(words), key=lambda x: x[0])

for key, group in groups:
    print(key, list(group))
```

**Python Data Classes**
n Python, the dataclass decorator is a feature introduced in Python 3.7 that provides a concise way to define classes primarily intended to store data. It automatically generates several special methods, such as init, repr, and eq, based on the class attributes you define. This simplifies the process of creating and working with data-focused classes


```python
from dataclasses import dataclass

@dataclass
class Person:
    name:str
    age:int
    profession:str
```

```python
person1=Person('ayu',17,'SE')
person1
```

```python
person1.age
```

```python
@dataclass(frozen=True)
class Point:
    x:int
    y:int
```

```python
point=Point(3,4)
```

```python
point.x,point.y
```

```python
## inheritance

@dataclass
class Person:
    name:str
    age:int

@dataclass
class Employee(Person):
    employee_id:str
    department:str
```
