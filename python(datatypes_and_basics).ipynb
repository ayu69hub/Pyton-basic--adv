```python
 #FACTORIAL


def fact(n):
  result =1
  for i in range(1,n+1):
    result = result*i

  return(result)

res = fact(27)
print(res)
```

```python
#HIGHER ORDER FUNCTION(jisme fun ke andr function ko call kiya jata h)
def square(n):
  return n*n

def cube(n):
  return n*n*n

def operate(n,operation):
  return operation(n)

print(operate(5,square))
print(operate(6,cube))
```

```python
#ANONYMOUS FUNCTION(jiska koi naam nii hota or isme lmbda use hota h)


square = lambda n:n*n
cube = lambda n:n*n*n
add = lambda a,b:a+b

result = square(5)
print(result)
result2 = cube(8)
print(result2)
result3 = add(5,6)
print(result3)
```

```python
#*RECURSION *


def fact(n):

  return n*fact(n-1) if n>1 else 1
fact(5)
```

```python
#FILTER

num = [2,3,4,5,6,7,8,9]
even = []

even = list(filter(lambda n:n%2==0,num))

print(even)
```

```python
#MAP
num = [2,3,4,5,6,7,8,9]

even = list(filter(lambda n:n%2==0,num))
double = list(map(lambda n:n*2,even))

print(double)
```

```python
#REDUCE

from functools import reduce
num = [2,3,4,5,6,7,8,9]

double = list(map(lambda n:n*2,even))
sum = reduce(lambda a,b:a+b,double)

print(sum)
```
